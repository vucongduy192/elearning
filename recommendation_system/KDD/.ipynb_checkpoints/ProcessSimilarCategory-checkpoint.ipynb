```python
import pandas as pd
```

```python
cat_df = pd.read_csv('./categories.csv')
course_df = pd.read_csv('./courses.csv')
```

```python
cat_df['category']
```

```python
total_courses = len(course_df['course_id'])
sim_C = {'course_id': [i for i in range(total_courses)]}
for i in range(total_courses):
    sim_C[i] = []
    for j in range(total_courses):
        cat1 = course_df[course_df['course_id'] == i].category.values[0]
        cat2 = course_df[course_df['course_id'] == j].category.values[0]
        sim_C[i].append(cat_df[cat_df['category'] == cat1][cat2].values[0])
```

```python
new_df = pd.DataFrame(sim_C)
new_df.head()
```

```python
new_df.to_csv('./simC_matrix.csv', index=False)
```
