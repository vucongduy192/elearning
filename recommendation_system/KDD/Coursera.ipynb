## Course detail

```python
def isEnglish(row):
    try:
        row['Name'].encode(encoding='utf-8').decode('ascii')
    except UnicodeDecodeError:
        return False
    else:
        return True
    
def hasRating(row):
    if row['Rating'] == "None":
        return False
    return True

def hasLevel(row):
    if row['Difficulty'] == "None":
        return False
    return True
```

```python
# Total records
df = pd.read_csv('coursera-course-detail-data.csv', 
                 converters={'Tags': lambda x: x[1:-1].replace("'", "").split(',')})
len(df)
```

```python
type(df.Rating[0])
```

```python
english_name = df.apply(isEnglish, axis=1)
df = df[english_name]
len(df)
```

```python
has_rating = df.apply(hasRating, axis=1)
df = df[has_rating]
len(df)
```

```python
hasLevel = df.apply(hasLevel, axis=1)
df = df[hasLevel]
len(df)
```
